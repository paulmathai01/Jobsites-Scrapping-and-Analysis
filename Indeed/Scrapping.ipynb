```python
URL = "https://www.indeed.com/jobs?q=data+science&start=10"
page = requests.get(URL)
soup = BeautifulSoup(page.text, "html.parser")
print(soup.prettify())
```

```python
def extract_job_title_from_result(soup): 
    jobs = []
    for div in soup.find_all(name="div", attrs={"class":"row"}):
      for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
          jobs.append(a["title"])
    return(jobs)
title = extract_job_title_from_result(soup)
title
```

```python
def extract_company_from_result(soup): 
    companies = []
    for div in soup.find_all(name="div", attrs={"class":"row"}):
        company = div.find_all(name="span", attrs={"class":"company"})
        if len(company) > 0:
            for b in company:
                companies.append(b.text.strip())
        else:
            sec_try = div.find_all(name="span", attrs={"class":"result-link-source"})
            for span in sec_try:
                companies.append(span.text.strip())
    return(companies)
 
company = extract_company_from_result(soup)
company
```

```python
def extract_location_from_result(soup): 
    locations = []
    spans = soup.findAll('span', attrs={'class': 'location'})
    for span in spans:
        locations.append(span.text)
    return(locations)
location = extract_location_from_result(soup)
location
```

```python
def extract_salary_from_result(soup): 
    salary = []
    spans = soup.findAll('span', attrs={'class': 'salaryText'})
    for span in spans:
        salary.append(span.text)
    return(salary)
salary = extract_salary_from_result(soup)
salary
```

```python

def extract_ratingsContent_from_result(soup): 
    ratingsContent = []
    spans = soup.findAll('span', attrs={'class': 'ratingsContent'})
    for span in spans:
        try:
            ratingsContent.append(span.text)
        except:
            ratingsContent.append("Nothing_found")
    return(ratingsContent)

ratingsContent = extract_ratingsContent_from_result(soup)
ratingsContent
```

```python
def extract_summary_from_result(soup): 
    summaries = []
    spans = soup.findAll('div', attrs={"class":"summary"})
    for span in spans:
        summaries.append()
    return(summaries)
extract_summary_from_result(soup)
```

```python
columns = ["job_title", "company_name", "location", "salary"]

sample_df = pd.DataFrame(columns = columns)

for start in range(0, 10000, 10):
    page = requests.get('http://www.indeed.com/jobs?q=data+science' + '&start=' + str(start))
    time.sleep(1)  #ensuring at least 1 second between page grabs
    soup = BeautifulSoup(page.text, "lxml", from_encoding="utf-8")
    for div in soup.find_all(name="div", attrs={"class":"row"}): 
        #specifying row num for index of job posting in dataframe
        num = (len(sample_df) + 1)
        job_post = [] 
        #grabbing job title
        for a in div.find_all(name="a", attrs={"data-tn-element":"jobTitle"}):
            job_post.append(a["title"]) 
        #grabbing company name
        company = div.find_all(name="span", attrs={"class":"company"}) 
        if len(company) > 0: 
            for b in company:
                job_post.append(b.text.strip()) 
        else: 
            sec_try = div.find_all(name="span", attrs={"class":"result-link-source"})
            for span in sec_try:
                job_post.append(span.text)
        #grabbing location name
        c = div.findAll("span", attrs={"class": "location"}) 
        for span in c: 
            job_post.append(span.text)
        #grabbing salary
        d = div.findAll('span', attrs={'class': 'ratingsContent'})
        for span in d:
            job_post.append(span.text)
        sample_df.loc[num] = job_post

sample_df.to_csv("data.csv", encoding='utf-8')
 
```
